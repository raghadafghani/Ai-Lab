```python
import matplotlib.pyplot as plt
import numpy as np
# generating a sinwwave signal
t = np.arange(0, 1, 0.001)
#t is an array created using np.arange(). It starts at 0, ends at 1 (exclusive),
# and increments by 0.001. This array represents time values.
sig = np.sin(2 * np.pi * 2 * t)
#sig is an array that stores the values of a sine wave signal.
#It's generated using np.sin() and is based on the time values in t. In this case, it's a sine wave with a frequency of 2 Hz (2 cycles per second).
plt.plot(t,sig)
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude')
plt.grid(True)
#adds a grid to the plot for better visualization.
plt.show()
```

```python
# generating two sinwwave signals
t = np.arange(0, 2, 0.001)
sig1 = np.sin(2 * np.pi * 2 * t)
sig2 = np.sin(2 * np.pi * 2 * t - np.pi/6)
sig3 = np.sin(2 * np.pi * 2 * t + np.pi/4)
#sig1, sig2, and sig3 are arrays that store the values of three different sine wave signals. sig1 is a sine wave with a frequency of 2 Hz,
#sig2 is a sine wave with the same frequency but phase-shifted by -pi/6 radians,
#and sig3 is a sine wave with the same frequency but phase-shifted by pi/4 radians.
fig, axs = plt.subplots(2, 1)
#This line creates a figure (fig) with two subplots arranged vertically. axs is a list of two axes objects, one for each subplot.

axs[0].plot(t, sig1)
axs[0].set_xlim(0, 2)
axs[0].set_xlabel('Time (sec)')
axs[0].set_ylabel('Amplitude')
axs[0].grid(True)



axs[1].plot(t, sig2, t, sig3)
axs[1].set_xlim(0, 2)
axs[1].set_xlabel('Time (sec)')
axs[1].set_ylabel('Amplitude')
axs[1].grid(True)
fig.tight_layout() #ensures that the subplots are properly spaced and don't overlap.
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate time values
t = np.arange(0, 2, 0.001)

# Generate two sine wave signals with different phases
sig1 = np.sin(2 * np.pi * 2 * t)
sig2 = np.sin(2 * np.pi * 2 * t - np.pi/6)

# Plotting the curves with markers and labels
plt.plot(t, sig1, label='Signal 1', linestyle='-', marker='o', markersize=4)
plt.plot(t, sig2, label='Signal 2', linestyle='--', marker='s', markersize=4)

# Adding a legend
plt.legend()

# Adding a title to the figure
plt.title('Sine Wave Signals')

# Adding labels to the axes
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude')

# Adding a grid
plt.grid(True)

# Display the figure
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate time values
t = np.arange(0, 2, 0.001)

# Generate sine and cosine wave signals
sin_wave = np.sin(2 * np.pi * 2 * t)
cos_wave = np.cos(2 * np.pi * 2 * t)

# Create a figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Plot the sine wave on the first subplot
ax1.plot(t, sin_wave, label='Sine Wave', linestyle='-', marker='o', markersize=4)

# Add a legend to the first subplot
ax1.legend()

# Set the title and labels for the first subplot
ax1.set_title('Sine Wave')
ax1.set_xlabel('Time (sec)')
ax1.set_ylabel('Amplitude')

# Add a grid to the first subplot
ax1.grid(True)

# Plot the cosine wave on the second subplot
ax2.plot(t, cos_wave, label='Cosine Wave', linestyle='--', marker='s', markersize=4)

# Add a legend to the second subplot
ax2.legend()

# Set the title and labels for the second subplot
ax2.set_title('Cosine Wave')
ax2.set_xlabel('Time (sec)')
ax2.set_ylabel('Amplitude')

# Add a grid to the second subplot
ax2.grid(True)

# Adjust the layout to prevent overlap
plt.tight_layout()

# Set a title for the entire figure
fig.suptitle('Sine and Cosine Waves')

# Display the figure
plt.show()
```

```python
# Compute percentiles using Pandas quantile() function
percentile_25 = df_cardio['height'].quantile(0.25)
percentile_50 = df_cardio['height'].quantile(0.5)
percentile_75 = df_cardio['height'].quantile(0.75)

print("25th Percentile:", percentile_25)
print("50th Percentile:", percentile_50)
print("75th Percentile:", percentile_75)

# Compute interquartile range (IQR)
iqr = percentile_75 - percentile_25
print("Interquartile Range (IQR):", iqr)

LowerBound_Height = percentile_25 - 1.5*iqr
UpperBound_Height = percentile_75 + 1.5*iqr
print(f"Lower Bound = {LowerBound_Height}, and Upper Bound = {UpperBound_Height}")
```
